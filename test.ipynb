```python
# Définition de la fonction qui renvoie un résultat Boolean (True ou False)
# indiquant si la phrase (sentence) est un pipotron ou non
def my_function(p_size: int, dict_possibilities: dict, sentence: str):
    # Séparer la phrase en une liste de composants avec la méthode split()
    components = sentence.split()

    # Vérifier si l'utilisateur a bien saisi la taille exacte du pipotron
    if len(components) != p_size:
        return False

    # Vérifier chaque composant par rapport aux ensembles de possibilités spécifiés dans le dictionnaire
    for i in range(p_size):
        # Si le composant actuel n'appartient pas à l'ensemble approprié, la phrase n'est pas un pipotron
        if components[i] not in dict_possibilities[i + 1]:
            return False

    # Si tous les composants sont conformes, la phrase est un pipotron
    return True

# Exemple d'application
p_size = 3
dict_possibilities = {
    1: ['Je', 'Tu', 'Demain je'],
    2: ['mange', 'bois', 'vole', 'voles'],
    3: ['.', '!', '?', '…']
}
sentence_1 = 'Je mange !'
sentence_2 = 'Tu dors ?'

# Appeler la fonction avec les paramètres spécifiés
result_1 = my_function(p_size, dict_possibilities, sentence_1)
result_2 = my_function(p_size, dict_possibilities, sentence_2)

# Afficher les résultats
print(result_1)  # True
print(result_2)  # False
```
